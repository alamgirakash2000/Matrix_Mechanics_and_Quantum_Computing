```python
# Importing Libraries
from qiskit import *
from qiskit.tools.monitor import  job_monitor
from qiskit.tools.visualization import plot_state_qsphere
from qiskit.visualization import plot_histogram, plot_bloch_multivector, array_to_latex
```

```python
# Create circuit
circuit = QuantumCircuit(1,1)    # For one bit

# Apply Gate
circuit.x(0)                   # Here 0 is the index of the qubits and x gate is a essex gate

# apply messurement
circuit.measure(0,0)

# Draw after Messurement
circuit.draw(initial_state = True, output='mpl')
```

## Run on a Simulator for statevector

```python
# Run on Simulator
simulator = Aer.get_backend('aer_simulator')     # Select Simulator
circuit.save_statevector()                       # Save State Vector
qobj = assemble(circuit)                         # Assemble
job = simulator.run(circuit)                     # run the simulation
job_monitor(job)                                 # Monitor Job
result = job.result()                            # Get Result

# View State vector output
print("printing State Vector")
final_state = result.get_statevector()     # Get StateVector    
plot_bloch_multivector(final_state)        # Plot statevector
```

## Custom State for single bit

```python
# Create circuit
circuit = QuantumCircuit(1,1)                     # Circuit for one bit
custom_state = '1'                                # Define state
circuit.initialize(custom_state, 0)               # Assign Custom state
circuit.measure(0,0)                              # apply messurement
circuit.draw(initial_state = True, output='mpl')  # Draw after Messurement
```

```python
# Run on Simulator
simulator = Aer.get_backend('aer_simulator')     # Select Simulator
circuit.save_statevector()                       # Save StateVector
qobj = assemble(circuit)                         # Assemble
job = simulator.run(circuit)                     # run the simulation
job_monitor(job)                                 # Monitor Job
result = job.result()                            # Get Result

# View State vector output
print("printing State Vector")
final_state = result.get_statevector()     # Get StateVector    
plot_bloch_multivector(final_state)        # Plot statevector
```

## Custom State for Multiple bit

```python
# Create circuit
circuit = QuantumCircuit(2,1)                     # Circuit for two bit
custom_state = '11'                               # Define state
circuit.initialize(custom_state, circuit.qubits)  # Assign Custom state
circuit.measure(0,0)                              # apply messurement
circuit.draw(initial_state = True, output='mpl')  # Draw after Messurement
```

```python
# Run on Simulator
simulator = Aer.get_backend('aer_simulator')     # Select Simulator
circuit.save_statevector()                       # Save StateVector
qobj = assemble(circuit)                         # Assemble
job = simulator.run(circuit)                     # run the simulation
job_monitor(job)                                 # Monitor Job
result = job.result()                            # Get Result

# View State vector output
print("printing StateVector")
final_state = result.get_statevector()     # Get StateVector    
plot_bloch_multivector(final_state)        # Plot statevector
```

## Qsphere Visualization

```python
plot_state_qsphere(final_state)
```

## Three Types of Ouput

### Unitary Matrix

```python
# Create circuit
circuit = QuantumCircuit(1,1)                     # Circuit for one bit
circuit.x(0)                                      # Apply Gate on 1st bit
circuit.draw(initial_state = True, output='mpl')  # Draw after Messurement
```

```python
# Run on Simulator for unitary matrix
simulator = Aer.get_backend('aer_simulator')     # Select Simulator
circuit.save_unitary()                           # Save Unitary Matrix
qobj = assemble(circuit)                         # Assemble
job = simulator.run(circuit)                     # run the simulation
job_monitor(job)                                 # Monitor Job
result = job.result()                            # Get Result

# View Unitary matrix
print("printing unitary Matrix")
unitary = result.get_unitary()          # Get Unitary     
print(unitary)                          # Print in array form
array_to_latex(unitary)                 # Print in matrix form
```

### Simulated Counts

```python
# Create circuit
circuit = QuantumCircuit(1,1)                     # Circuit for one bit
circuit.h(0)                                      # Apply Hadamard Gate on 1st bit
circuit.measure(0,0)
circuit.draw(initial_state = True, output='mpl')  # Draw after Messurement
```

```python
# Run on Simulator
simulator = Aer.get_backend('aer_simulator')     # Select Simulator
circuit.save_statevector()                       # Save stateVector
qobj = assemble(circuit)                         # Assemble
job = simulator.run(circuit)                     # run the simulation
job_monitor(job)                                 # Monitor Job
result = job.result()                            # Get Result

# Simulated counts output
counts = result.get_counts()       # Get counts
plot_histogram(counts)             # plot counts
```

## Two Qubit system

```python
# Create circuit
circuit = QuantumCircuit(2,1)                # Circuit for two bit
circuit.x(0)                                 # First qubit will be 1
circuit.cx(0,1)                              # Apply cx gate on both bits
circuit.measure(0,0)
circuit.draw(initial_state = True, output='mpl')  # Draw after Messurement
```

```python
# Run on Simulator
simulator = Aer.get_backend('aer_simulator')     # Select Simulator
circuit.save_statevector()                       # Save StateVector
qobj = assemble(circuit)                         # Assemble
job = simulator.run(circuit)                     # run the simulation
job_monitor(job)                                 # Monitor Job
result = job.result()                            # Get Result

# View State vector output
print("printing StateVector")
final_state = result.get_statevector()     # Get StateVector    
plot_bloch_multivector(final_state)        # Plot statevector
```

**Comment:** As default bit was 00, `x` gate change 1st 0 to 1. For that 2nd bit(control bit) become 1 after passing through `cx` gate.
